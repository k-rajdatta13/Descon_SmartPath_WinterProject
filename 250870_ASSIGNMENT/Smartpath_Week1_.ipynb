Task 4:

```python
import pandas as pd
Traffic_data = pd.DataFrame({
    "Road_ID" : range(1,len(speed_array)+1),
    "Speed" : speed_array,
    "Length" : [2]*len(speed_array)
})
```

```python
Traffic_data["Travel_Time"] = (Traffic_data["Length"]/Traffic_data["Speed"])*60
```

```python
Sorted_Traffic_data = Traffic_data.sort_values("Travel_Time")
print(Sorted_Traffic_data)
```

```python
Sorted_Traffic_data.loc[Sorted_Traffic_data["Speed"] < 10, "Traffic_Level"] = "Very High"
Sorted_Traffic_data.loc[(Sorted_Traffic_data["Speed"] >= 10) & (Sorted_Traffic_data["Speed"] < 30), "Traffic_Level"] = "High"
Sorted_Traffic_data.loc[(Sorted_Traffic_data["Speed"] >= 30) & (Sorted_Traffic_data["Speed"] < 50), "Traffic_Level"] = "Moderate"
Sorted_Traffic_data.loc[Sorted_Traffic_data["Speed"] >= 50, "Traffic_Level"] = "Low"
print(Sorted_Traffic_data)
```

Task 5:

```python
import matplotlib.pyplot as plt
plt.plot(Sorted_Traffic_data["Road_ID"],Sorted_Traffic_data["Speed"])
plt.xlabel("Road ID")
plt.ylabel("Speed")
plt.title("Plot of Speed vs Road ID ")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.bar(Sorted_Traffic_data["Road_ID"],Sorted_Traffic_data["Travel_Time"])
plt.xlabel("Road ID")
plt.ylabel("Travel Time")
plt.title("Plot of Travel Time vs Road ID ")
plt.grid(True)
plt.show()
```
